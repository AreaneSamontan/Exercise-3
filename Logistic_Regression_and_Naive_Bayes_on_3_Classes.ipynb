```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB 
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, cross_val_predict
from utilities import visualize_classifier
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Data/three-classes.csv')
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values

# Separate input data into two classes based on labels
class_0 = np.array(X[y==0])
class_1 = np.array(X[y==1])
class_2 = np.array(X[y==2])
```

```python
plt.figure()
plt.scatter(class_0[:, 0], class_0[:, 1], s=75, facecolors='yellow',
 edgecolors='black', linewidth=1, marker='o')
plt.scatter(class_1[:, 0], class_1[:, 1], s=75, facecolors='red',
 edgecolors='black', linewidth=1, marker='o')
plt.scatter(class_2[:, 0], class_2[:, 1], s=75, facecolors='green',
 edgecolors='black', linewidth=1, marker='o')
plt.title('Input data')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

#Naive Bayes Performance

```python
classifierNB = GaussianNB()
classifierNB.fit(X_train, y_train)
y_test_pred = classifierNB.predict(X_test)
```

```python
visualize_classifier(classifierNB, X_test, y_test)
```

```python
accuracy = 100.0 * (y_test == y_test_pred).sum() / X_test.shape[0]
print("Accuracy of the NB classifier =", round(accuracy, 2), "%")
```

#Logistic Regression Performance

```python
classifierLR = linear_model.LogisticRegression(solver='liblinear', C=100)
```

```python
classifierLR.fit(X_train, y_train)
```

```python
y_test_predLR = classifierLR.predict(X_test)
```

```python
accuracyLR = 100.0 * (y_test == y_test_predLR).sum() / X_test.shape[0]
print("Accuracy of the LR classifier =", round(accuracyLR, 2), "%")
```

```python
visualize_classifier(classifierLR, X_test, y_test)
```

#CONCLUSION:
#####Naive Bayes and Logistic Regression have good results because the data clusters are not overlapped
